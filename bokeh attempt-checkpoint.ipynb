```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
!pip install seaborn
import seaborn as sns

# Bokeh libraries
!pip install bokeh
from bokeh.io import output_file, output_notebook
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.layouts import row, column, gridplot
from bokeh.models.widgets import Tabs, Panel
from bokeh.charts import Bar, output_file, show
```

```python
df = pd.read_csv('https://query.data.world/s/5dpnyov52z2lsghtptp4v6n473tm6l')
newdf=pd.melt(df, id_vars = 'Item',var_name='Year', value_name ='No of Transactions')
newdf["No of Transactions"]=newdf["No of Transactions"]/1000000
newdf=newdf.round(1)
newdf
indexNames = newdf[newdf['Item'] == "TOTAL"].index 
newdf.drop(indexNames , inplace=True)
print(newdf.info())
onlyv = newdf[newdf['Item'].str.contains('ehicle')]
print (onlyv)
indexNames = onlyv[onlyv['Item'] == "791--Railway Vehicles & Associated Equipment"].index 
onlyv.drop(indexNames , inplace=True)
print(onlyv)
```

```python
# The figure will be rendered in a static HTML file called output_file_test.html
output_file('output_file_test.html', 
            title='Empty Bokeh Figure')
output_notebook()  # Render inline in a Jupyter Notebook

# Set up a generic figure() object
fig = figure()

# See what it looks like
show(fig)
```

```python
cds = ColumnDataSource(onlyv)  #https://docs.bokeh.org/en/latest/docs/user_guide/data.html
cds
```

```python
from bokeh.charts import Bar, output_file, show
#from bokeh.sampledata.autompg import autompg as df

#http://docs.bokeh.org/en/0.11.0/docs/user_guide/concepts.html  bokeh charts not supported in some notebooks

bokehplot = Bar(onlyv, 'No of Transactions', values='Year', title="US-DE Balances")

output_file("bar.html")

show(bokehplot)
```

```python
fig = figure(background_fill_color='gray',
             background_fill_alpha=0.5,
             border_fill_color='blue',
             border_fill_alpha=0.25,
             plot_height=300,
             plot_width=500,
             h_symmetry=True,
             x_axis_label='No of Transactions',
             x_axis_type='datetime',
             x_axis_location='above',
             x_axis=cds["Item",
             y_axis_label='Year',
             y_axis_type='linear',
             y_axis_location='right',
             y_axis=cds.onlyv["Year"],
             title='Example Figure',
             title_location='right',
             toolbar_location='below',
             tools='save')

#graphoutlegend=sns.barplot(x ='No of Transactions', y ='Year', hue='Item', data = onlyv, palette ='plasma')
```
